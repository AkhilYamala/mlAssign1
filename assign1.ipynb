```
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 
```

```
data = pd.read_csv('housing_data.csv')
```

```
def using_different_attributes(attribute):
  X = np.array(data[attribute])
  y = np.array(data['price'])
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
  X_train = X_train.reshape(-1, 1)
  X_test = X_test.reshape(-1, 1)
  reg = LinearRegression()
  reg.fit(X_train, y_train)
  print('Training score of',attribute, 'is',reg.score(X_train, y_train))
  print('Testing score of',attribute, 'is',reg.score(X_test, y_test))
  plot(reg, attribute, X_train, y_train)
```

```
def plot(reg, attribute, X, y):
  plt.scatter(X, y)
  plt.plot(X, reg.predict(X), color='red', linewidth=3)
  plt.xlabel(attribute)
  plt.ylabel('price')
  plt.show()
```

```
def get_attributes(data):
  attributes = []
  for col in data.columns:
    if type(data[col][0]) != str and col != 'price':
      attributes.append(col)
  return attributes
```

**Using different attributes**



```
attributes = get_attributes(data)
for attribute in attributes:
  using_different_attributes(attribute)
```

**Multi variant linear regression**

```
# multi variant linear regression
X = np.array(data[attributes])
y = np.array(data['price'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
reg = LinearRegression()
reg.fit(X_train,y_train)
print(reg.score(X_test, y_test))
```
